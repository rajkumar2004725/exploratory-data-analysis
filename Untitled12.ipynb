18. Derive a new column called agegroup from age column where persons falling in different age ranges
are categorized as below.
(015): young
(1535): adults
(3555): mid
(55): old


```python
def mapage(age:int):
  if age>=0 and age<15:
    return 'young'
  elif age>=15 and age<35:
    return 'adults'
  elif age>=35 and age<55:
    return 'mid'
  else:
    return 'old'
df['agegroup']=df['age'].values.tolist()
df['agegroup']=df['agegroup'].apply(mapage)
```

```python
df.agegroup.value_counts()
```
